```python
for i in Data_Indo_Jkt.columns:  
    Q1  = Data_Indo_Jkt[i].quantile(0.25)
    Q3  = Data_Indo_Jkt[i].quantile(0.75)
    IQR = Q3-Q1

    Lower_Bound = Q1-(1.5*IQR)
    Upper_Bound = Q3+(1.5*IQR)

    print('Kolom',i,'Memiliki Outliers sebesar:',len(Data_Indo_Jkt[(Data_Indo_Jkt[i]>Upper_Bound)|(Data_Indo_Jkt[i]<Lower_Bound)])/len(Data_Indo_Jkt)*100,'%')
print('\n')
```
